```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

train_df = pd.read_csv('train.csv')
oil_df = pd.read_csv('oil.csv')
stores_df = pd.read_csv('stores.csv')
holidays_df = pd.read_csv('holidays_events.csv')
transactions_df = pd.read_csv('transactions.csv')

print("Tüm veriler başarıyla yüklendi!")
```

```python
oil_df['date'] = pd.to_datetime(oil_df['date'])

oil_df = oil_df.set_index('date').asfreq('D')

# Boş günleri bir önceki günün petrol fiyatıyla dolduruyoruz (Forward Fill)
oil_df['dcoilwtico'] = oil_df['dcoilwtico'].ffill()

# Eğer en baştaki gün boş kaldıysa, onu da bir sonraki günün fiyatıyla dolduruyoruz (Backward Fill)
oil_df['dcoilwtico'] = oil_df['dcoilwtico'].bfill()

oil_df = oil_df.reset_index()

print(oil_df.isnull().sum())
```

```python
daily_sales = merged_df.groupby('date')['sales'].sum().reset_index()

plt.figure(figsize=(15, 6))

sns.lineplot(data=daily_sales, x='date', y='sales', color='royalblue', linewidth=1.5)

plt.title('Ekvador Süpermarket Zinciri Toplam Günlük Satış Trendi (2013 - 2017)', fontsize=14, fontweight='bold')
plt.xlabel('Tarih', fontsize=12)
plt.ylabel('Toplam Satış Adedi', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6) 

plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
daily_sales_idx = daily_sales.set_index('date')

result = seasonal_decompose(daily_sales_idx['sales'], model='additive', period=365)

plt.figure(figsize=(15, 12))

plt.subplot(4, 1, 1)
plt.plot(result.observed, color='royalblue')
plt.title('Orijinal Günlük Satış Verisi', fontsize=12, fontweight='bold')
plt.grid(True, linestyle='--')

plt.subplot(4, 1, 2)
plt.plot(result.trend, color='darkorange', linewidth=2)
plt.title('Sadece Trend (Uzun Vadeli Büyüme)', fontsize=12, fontweight='bold')
plt.grid(True, linestyle='--')

plt.subplot(4, 1, 3)
plt.plot(result.seasonal, color='forestgreen')
plt.title('Sadece Mevsimsellik (Yıllık Döngü)', fontsize=12, fontweight='bold')
plt.grid(True, linestyle='--')

plt.subplot(4, 1, 4)
plt.scatter(result.resid.index, result.resid, color='crimson', s=3, alpha=0.5)
plt.title('Artıklar (Residuals / Beklenmedik Şoklar)', fontsize=12, fontweight='bold')
plt.grid(True, linestyle='--')

plt.tight_layout()
plt.show()
```

```python
####### FEATURE ENGINEERING #######
```

```python
merged_df['sales_lag_1'] = merged_df.groupby(['store_nbr', 'family'])['sales'].shift(1)

merged_df['sales_lag_2'] = merged_df.groupby(['store_nbr', 'family'])['sales'].shift(2)

merged_df['sales_lag_7'] = merged_df.groupby(['store_nbr', 'family'])['sales'].shift(7)

merged_df['sales_lag_14'] = merged_df.groupby(['store_nbr', 'family'])['sales'].shift(14)

merged_df[['date', 'store_nbr', 'family', 'sales', 'sales_lag_1', 'sales_lag_7']].dropna().head()
```

```python
merged_df['sales_roll_mean_7'] = merged_df.groupby(['store_nbr', 'family'])['sales'].shift(1).transform(lambda x: x.rolling(7).mean())

merged_df['sales_roll_mean_30'] = merged_df.groupby(['store_nbr', 'family'])['sales'].shift(1).transform(lambda x: x.rolling(30).mean())

merged_df['sales_roll_std_7'] = merged_df.groupby(['store_nbr', 'family'])['sales'].shift(1).transform(lambda x: x.rolling(7).std())

merged_df[['date', 'sales', 'sales_roll_mean_7', 'sales_roll_mean_30']].dropna().head()
```

```python
merged_df['month'] = merged_df['date'].dt.month          
merged_df['day_of_month'] = merged_df['date'].dt.day  
merged_df['is_month_end'] = merged_df['date'].dt.is_month_end.astype(int) 
merged_df['is_month_start'] = merged_df['date'].dt.is_month_start.astype(int)

merged_df['is_weekend'] = merged_df['day_of_week'].apply(lambda x: 1 if x >= 5 else 0)

merged_df[['date', 'month', 'day_of_month', 'is_month_end', 'is_weekend']]
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
categorical_cols = ['family', 'city', 'state', 'type']
le = LabelEncoder()

for col in categorical_cols:
    merged_df[col] = le.fit_transform(merged_df[col])

display(merged_df.head())
```

```python
###### TRAIN TEST ######
```

```python
df_clean = merged_df.dropna().copy()

split_date = '2017-07-01'

train_data = df_clean[df_clean['date'] < split_date]
val_data = df_clean[df_clean['date'] >= split_date]

drop_cols = ['id', 'date', 'sales', 'day_name', 'has_promotion']
features = [col for col in df_clean.columns if col not in drop_cols]

X_train = train_data[features]
y_train = train_data['sales']

X_val = val_data[features]
y_val = val_data['sales']

print(f"Eğitim Seti Boyutu: {X_train.shape}")
print(f"Doğrulama (Validation) Seti Boyutu: {X_val.shape}")
print("\nModelin kullanacağı canavar özellikler (Features):\n", features)
```

```python
##### MODEL DENEMELERİ #####
```

```python
merged_df['is_holiday'] = merged_df['is_holiday'].map({'Tatil Günü': 1, 'Normal Gün': 0})

df_clean = merged_df.dropna().copy()

train_data = df_clean[df_clean['date'] < split_date]
val_data = df_clean[df_clean['date'] >= split_date]

X_train = train_data[features]
y_train = train_data['sales']

X_val = val_data[features]
y_val = val_data['sales']

```

```python
import lightgbm as lgb
from sklearn.metrics import root_mean_squared_error

lgb_model = lgb.LGBMRegressor(
    n_estimators=500,
    learning_rate=0.05,
    random_state=42,
    n_jobs=-1
)

lgb_model.fit(X_train, y_train)

lgb_preds = lgb_model.predict(X_val)
lgb_preds = np.clip(lgb_preds, 0, None)

lgb_rmse = root_mean_squared_error(y_val, lgb_preds)
print(f"\nLightGBM Validation RMSE Skoru: {lgb_rmse:.4f}")
```

```python
import xgboost as xgb

xgb_model = xgb.XGBRegressor(
    n_estimators=500,
    learning_rate=0.05,
    random_state=42,
    n_jobs=-1,
    tree_method='hist'
)

xgb_model.fit(X_train, y_train)

xgb_preds = xgb_model.predict(X_val)
xgb_preds = np.clip(xgb_preds, 0, None)

xgb_rmse = root_mean_squared_error(y_val, xgb_preds)
print(f"\n XGBoost Validation RMSE Skoru: {xgb_rmse:.4f}")
```

```python

val_analysis = val_data.copy()
val_analysis['lgb_preds'] = lgb_preds 

daily_actual = val_analysis.groupby('date')['sales'].sum().reset_index()
daily_preds = val_analysis.groupby('date')['lgb_preds'].sum().reset_index()

plt.figure(figsize=(15, 6))

sns.lineplot(data=daily_actual, x='date', y='sales', color='royalblue', label='Gerçek Satışlar (Actual)', linewidth=2)

sns.lineplot(data=daily_preds, x='date', y='lgb_preds', color='darkorange', label='Model Tahminleri (Predicted)', linewidth=2, linestyle='--')

plt.title('Gelecek Tahmini Kontrolü: Gerçek Satışlar vs LightGBM Tahminleri', fontsize=14, fontweight='bold')
plt.xlabel('Tarih (Validation Dönemi)', fontsize=12)
plt.ylabel('Toplam Günlük Satış Adedi', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(fontsize=12)

plt.show()
```

```python
val_analysis_xgb = val_data.copy()
val_analysis_xgb['xgb_preds'] = xgb_preds

daily_actual_xgb = val_analysis_xgb.groupby('date')['sales'].sum().reset_index()
daily_preds_xgb = val_analysis_xgb.groupby('date')['xgb_preds'].sum().reset_index()

plt.figure(figsize=(15, 6))

sns.lineplot(data=daily_actual_xgb, x='date', y='sales', color='royalblue', label='Gerçek Satışlar (Actual)', linewidth=2)

sns.lineplot(data=daily_preds_xgb, x='date', y='xgb_preds', color='forestgreen', label='XGBoost Tahminleri (Predicted)', linewidth=2, linestyle='--')

plt.title('Gelecek Tahmini Kontrolü: Gerçek Satışlar vs XGBoost Tahminleri', fontsize=14, fontweight='bold')
plt.xlabel('Tarih (Validation Dönemi)', fontsize=12)
plt.ylabel('Toplam Günlük Satış Adedi', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(fontsize=12)

plt.show()
```

```python
from catboost import CatBoostRegressor
from sklearn.metrics import root_mean_squared_error


cat_model = CatBoostRegressor(
    iterations=500,
    learning_rate=0.05,
    depth=6,
    random_seed=42,
    verbose=100, 
    thread_count=-1
)

print("CatBoost eğitiliyor...")
cat_model.fit(X_train, y_train)

cat_preds = cat_model.predict(X_val)
cat_preds = np.clip(cat_preds, 0, None)

cat_rmse = root_mean_squared_error(y_val, cat_preds)
print(f"\n🎯 Sadece CatBoost Validation RMSE Skoru: {cat_rmse:.4f}")
```

```python
from sklearn.ensemble import RandomForestRegressor

rf_model = RandomForestRegressor(
    n_estimators=100,
    max_depth=10,
    random_state=42,
    n_jobs=-1, 
    verbose=1
)

print("Random Forest eğitiliyor")
rf_model.fit(X_train, y_train)

rf_preds = rf_model.predict(X_val)
rf_preds = np.clip(rf_preds, 0, None)

rf_rmse = root_mean_squared_error(y_val, rf_preds)
print(f"\n🎯 Sadece Random Forest Validation RMSE Skoru: {rf_rmse:.4f}")
```

```python
from prophet import Prophet

prophet_train = train_data[['date', 'sales']].rename(columns={'date': 'ds', 'sales': 'y'})
prophet_val = val_data[['date', 'sales']].rename(columns={'date': 'ds', 'sales': 'y'})

prophet_model = Prophet(yearly_seasonality=True, weekly_seasonality=True, daily_seasonality=False)
print("Prophet modeli eğitiliyor...")
prophet_model.fit(prophet_train)

forecast = prophet_model.predict(prophet_val[['ds']])

prophet_preds = np.clip(forecast['yhat'].values, 0, None)
prophet_rmse = root_mean_squared_error(prophet_val['y'], prophet_preds)

print(f"\n Sadece Prophet Validation RMSE Skoru: {prophet_rmse:.4f}")
```

```python
import pandas as pd

model_results = {
    "Model Adı": [
        "LightGBM Regressor",
        "XGBoost Regressor",
        "CatBoost Regressor",
        "Random Forest",
        "Prophet (Meta)"
    ],
    "Validation RMSE": [
        214.2243,  
        215.2127,  
        cat_rmse,   
        rf_rmse,    
        prophet_rmse 
    ]
}

df_comparison = pd.DataFrame(model_results)

df_comparison = df_comparison.sort_values(by="Validation RMSE", ascending=True).reset_index(drop=True)

print("=" * 50)
print("       🏆 MODEL PERFORMANS KARŞILAŞTIRMASI       ")
print("=" * 50)
print(df_comparison.to_string(index=False))
print("=" * 50)
print(f"🥇 En Başarılı Model: {df_comparison.iloc[0]['Model Adı']} (RMSE: {df_comparison.iloc[0]['Validation RMSE']:.4f})")
print("=" * 50)
```
